# Flats Constructed (Data)

```python
title = "Flat Constructed"
titlelen = len(title)
print("{:*^{titlelen}}".format(title, titlelen=titlelen+6))
print()

data = np.genfromtxt('DataProject/flats-constructed-by-housing-and-development-board-annual.csv',  
                               skip_header=1, 
                               dtype=[('year','U4'),  ('flat_constructed','i4')], delimiter=",",
                               missing_values=['na'],filling_values=0)


print("There are {} rows in this dataset".format(len(data)))
print()

set_year = set(data['year']) 
set_flat_constructed = set(data['flat_constructed'])


print(str(len(set_year)) + " unique values in year column")
print(str(len(set_flat_constructed)) + " unique values in flat_constructed column")

max_value = np.max(data['flat_constructed'])
min_value = np.min(data['flat_constructed'])         

max_row_index = np.argmax(data['flat_constructed'])
min_row_index = np.argmin(data['flat_constructed'])

std = np.std(data['flat_constructed'])
var = np.var(data['flat_constructed'])

print()

print("The mean number of flat constructed over the years is {:.0f}".format(np.mean(data['flat_constructed'])))

print("The maximum number of flat constructed was in the year {} is {}".format(data[max_row_index][0],data[max_row_index][1]))
print("The minimum number of flat constructed was in the year {} is {}".format(data[min_row_index][0],data[min_row_index][1]))

print("The standard deviation is {:.0f}".format(std))
print("The Variance is {:.0f}".format(var))
```

```python
title = "Number of Sold and Rented Residential Units"
titlelen = len(title)
print("{:*^{titlelen}}".format(title, titlelen=titlelen+6))
print()

data = np.genfromtxt('DataProject/number-of-sold-and-rented-residential-units.csv',  
                skip_header=1, 
                dtype=[('year','i4'),  ('property_type','U4'), ('sold_or_rented','U6'), ('flat_type','U50'), 
                       ('no_of_units','i4')], delimiter=",",
                missing_values=['na','-'],filling_values=0)

#Never sort yet
print("There are {} rows in this dataset".format(len(data)))
print()

set_year = set(data['year'])
set_property_type = set(data['property_type'])
set_sold_or_rented = set(data['sold_or_rented'])
set_flat_type = set(data['flat_type'])
set_no_of_units = set(data['no_of_units'])

print(str(len(set_year)) + " unique values in year column")
print(str(len(set_property_type)) + " unique values in property_type column")
print(str(len(set_sold_or_rented)) + " unique values in sold_or_rented column")
print(str(len(set_flat_type)) + " unique values in flat_type column")
print(str(len(set_no_of_units)) + " unique values in no_of_units column")

max_value = np.max(data['no_of_units'])
max_row_index = np.argmax(data['no_of_units'])

 
std = np.std(data['no_of_units'])
var = np.var(data['no_of_units'])

print()

print("The mean number of flat sold/rented over the years is {:.0f}".format(np.mean(data['no_of_units'])))
print("The maximum number of flat sold/rented was in the year {} is {} on {}".format(data[max_row_index][0],
                                                                                     data[max_row_index][4],
                                                                                     data[max_row_index][3]))
print("The standard deviation of flat sold/rented is {:.0f}".format(std))
print("The Variance of flat sold/rented is  {:.0f}".format(var))

#Sorted Rented
print()
def rented(data):
    data_rented = data[data['sold_or_rented']=='rented']
    return data_rented

max_value = np.max(rented(data)['no_of_units'])
max_row_index = np.argmax(rented(data)['no_of_units'])

 
std = np.std(rented(data)['no_of_units'])
var = np.var(rented(data)['no_of_units'])

print()

print("The mean number of flat rented over the years is {:.0f}".format(np.mean(rented(data)['no_of_units'])))
print("The maximum number of flat rented was in the year {} is {} on {}".format(rented(data)[max_row_index][0],
                                                                                rented(data)[max_row_index][4],
                                                                                rented(data)[max_row_index][3]))
print("The standard deviation of flat rented is {:.0f}".format(std))
print("The Variance of flat rented is {:.0f}".format(var))

#Sorted Sold
def sold(data):
    data_sold = data[data['sold_or_rented']=='sold']
    return data_sold



max_value = np.max(sold(data)['no_of_units'])
max_row_index = np.argmax(sold(data)['no_of_units'])

 
std = np.std(sold(data)['no_of_units'])
var = np.var(sold(data)['no_of_units'])

print()

print("The mean number of flat sold over the years is {:.0f}".format(np.mean(sold(data)['no_of_units'])))
print("The maximum number of flat sold was in the year {} is {} on {}".format(sold(data)[max_row_index][0],
                                                                              sold(data)[max_row_index][4],
                                                                              sold(data)[max_row_index][3]))
print("The standard deviation of flat sold is {:.0f}".format(std))
print("The Variance of flat sold is {:.0f}".format(var))
```

# 1990 - 2018 Data on Resale Flat Price (Data)

```python
title = "Resale Flat Prices"
titlelen = len(title)
print("{:*^{titlelen}}".format(title, titlelen=titlelen+6))
print()

data_1 = np.genfromtxt('DataProject/Resale/1resale-flat-prices-based-on-approval-date-1990-1999.csv',  
                skip_header=1, 
                dtype=[('date','U7'),  ('town','U50'), ('flat_type','U50'), 
                       ('block','U50'), ('street_name','U50'),('storey_range','U50'),
                       ('floor_area_sqm','i4'),('flat_model','U50'),('lease_commence_date','i4'),
                       ('resale_price','i4')], delimiter=",",
                        missing_values=['na'],filling_values=0)

data_2 = np.genfromtxt('DataProject/Resale/2resale-flat-prices-based-on-approval-date-2000-feb-2012.csv',  
                skip_header=1, 
                dtype=[('date','U7'),  ('town','U50'), ('flat_type','U50'), 
                       ('block','U50'), ('street_name','U50'),('storey_range','U50'),
                       ('floor_area_sqm','i4'),('flat_model','U50'),('lease_commence_date','i4'),
                       ('resale_price','i4')], delimiter=",",
                        missing_values=['na'],filling_values=0)
data_3 = np.genfromtxt('DataProject/Resale/3resale-flat-prices-based-on-registration-date-from-mar-2012-to-dec-2014.csv',  
                skip_header=1, 
                dtype=[('date','U7'),  ('town','U50'), ('flat_type','U50'), 
                       ('block','U50'), ('street_name','U50'),('storey_range','U50'),
                       ('floor_area_sqm','i4'),('flat_model','U50'),('lease_commence_date','i4'),
                       ('resale_price','i4')], delimiter=",",
                        missing_values=['na'],filling_values=0)

#special got remaining lease unlike others resale flat data, so i have to remove remaining_lease column
ignore=[9] #9 = remaining_lease column
data_4_wanted_column = [ind for ind in range(11) if ind not in ignore] #Not counting removing remaining_lease column
data_4 = np.genfromtxt('DataProject/Resale/4resale-flat-prices-based-on-registration-date-from-jan-2015-onwards.csv',  
                skip_header=1, 
                dtype=[('date','U7'),  ('town','U50'), ('flat_type','U50'), 
                       ('block','U50'), ('street_name','U50'),('storey_range','U50'),
                       ('floor_area_sqm','i4'),('flat_model','U50'),('lease_commence_date','i4'),
                       ('resale_price','i4')], delimiter=",",usecols = (data_4_wanted_column),
                        missing_values=['na'],filling_values=0)

data = np.concatenate([data_1,data_2,data_3,data_4])

#there is 2 type of MULTI-GENERATION/MULTI GENERATION
for i in range(len(data)):
    if(data['flat_type'][i]== 'MULTI-GENERATION'):
        data['flat_type'][i] = 'MULTI GENERATION'


set_date               = set(data['date'])
set_town               = set(data['town'])
set_flat_type          = set(data['flat_type']) 
set_block              = set(data['block'])
set_street_name        = set(data['street_name'])
set_storey_range       = set(data['storey_range'])           
set_floor_area_sqm     = set(data['floor_area_sqm'])           
set_flat_model         = set(data['flat_model'])           
set_lease_commence_date= set(data['lease_commence_date'])           
set_resale_price       = set(data['resale_price'])  

print("There are {} rows in this dataset".format(len(data)))

#Towns
town = np.array(list(set_town))
town.sort()

print()
         
print(str(len(set_date               )) + " unique values in date column")
print(str(len(set_town               )) + " unique values in town column")
print(str(len(set_flat_type          )) + " unique values in flat_type column")
print(str(len(set_block              )) + " unique values in block column")
print(str(len(set_street_name        )) + " unique values in street_name column")
print(str(len(set_storey_range       )) + " unique values in storey_range column")
print(str(len(set_floor_area_sqm     )) + " unique values in floor_area_sqm column")
print(str(len(set_flat_model         )) + " unique values in flat_model column")
print(str(len(set_lease_commence_date)) + " unique values in lease_commence_date column")
print(str(len(set_resale_price       )) + " unique values in resale_price column")


#Unique Year
resale_year = []
for i in range(len(set_date)):
    resale_year.append(int(list(set_date)[i][0:4]))
resale_year = np.array(resale_year)
set_resale_year = set(resale_year)
print("There is {} unique years ".format(str(len(set_resale_year))))


max_value = np.max(data['resale_price'])
min_value = np.min(data['resale_price'])  

max_row_index = np.argmax(data['resale_price'])
min_row_index = np.argmin(data['resale_price'])
 
std = np.std(data['resale_price'])
var = np.var(data['resale_price'])

print()

print("The mean price of Resale Flat Prices over the years is ${:.2f}".format(np.mean(data['resale_price'])))
print("The maximum price of Resale Flat Prices was in year {} is ${} at {}".format(data[max_row_index][0],data[max_row_index][-1],data[max_row_index][1]))
print("The minimum price of Resale Flat Prices was in year {} is ${} at {}".format(data[min_row_index][0],data[min_row_index][-1],data[min_row_index][1]))
print("The standard deviation of Resale Flat Prices is ${:.2f}".format(std))
print("The Variance of Resale Flat Prices is  ${:.2f}".format(var))
```
